```python
#Looking at all the categories available in the loan status to understand where to focus and correlate
df["loan_status"].unique()
```

```python
#This dataset is huge. I will create a subset to work with the things required. Creating a subset copy of the original data
focus_cols = ["loan_amnt","funded_amnt","term","int_rate","installment","grade","sub_grade",
    "annual_inc","emp_length","home_ownership","purpose","dti",
    "fico_range_low","fico_range_high","revol_util",
    "open_acc","total_acc","loan_status"]

#making sure that the above defined columns for subset are present in main data

existing_focus = [c for c in focus_cols if c in df.columns]
print(existing_focus)

#Creating a copy with the needed data

dfcopy = df[existing_focus].copy()

print(dfcopy.head())
```

```python
#Checking the number of defaulters and other categories
print(dfcopy["loan_status"].value_counts())
```

```python

#Only keeping the useful categories from loan_status. Dropping the rest.
default_like = {"Charged Off", "Default"}
paid_like = {"Fully Paid"}

mask_keep = dfcopy["loan_status"].isin(default_like.union(paid_like))
dfcopy = dfcopy[mask_keep].copy()
```

```python
#Focusing on two major categories
dfcopy["default_flag"] = np.where(dfcopy["loan_status"].isin(default_like), 1,0)
print(dfcopy["default_flag"].value_counts())
```

```python
#checking the percentage of defaulters and good customers
print(dfcopy["default_flag"].value_counts(normalize=True))
```

```python
#Time to clean data. Covert numbers and strings to original.
if "term" in dfcopy.columns and dfcopy["term"].dtype == object:
    dfcopy["term_months"] = dfcopy["term"].str.extract(r"(\d+)").astype(float)


if "int_rate" in dfcopy.columns and dfcopy["int_rate"].dtype == object:
    dfcopy["int_Rate_perc"] = dfcopy["int_rate"].str.rstrip("%").astype(float)

if "revol_util" in dfcopy.columns and dfcopy["revol_util"].dtype == object:
    dfcopy["revol_util_perc"] = dfcopy["revol_util"].str.rstrip("%").replace("",np.nan).astype(float)

if "emp_length" in dfcopy.columns and dfcopy["emp_length"].dtype == object:
    emp_map = {"10+ years": 10, "9 years": 9, "8 years": 8, "7 years": 7, "6 years": 6,
        "5 years": 5, "4 years": 4, "3 years": 3, "2 years": 2, "1 year": 1,
        "< 1 year": 0, "n/a": np.nan}
    dfcopy["emp_lenght_years"] = dfcopy["emp_length"].map(emp_map)


if "fico_range_low" in dfcopy.columns and "fico_range_high" in dfcopy.columns:
    dfcopy["fico_mid"] = (dfcopy["fico_range_low"] + dfcopy["fico_range_high"]) / 2.0
```

```python
#Descriptive statistics for key numeric features
candidate_num = [ "loan_amnt", "funded_amnt", "int_rate_clean", "annual_inc", "dti",
    "fico_mid", "revol_util_clean", "open_acc", "total_acc", "term_clean", "int_rate", "total_pymnt"]

num_cols = [c for c in candidate_num if c in dfcopy.columns]
print(dfcopy[num_cols].describe().T)
```

```python
#Class balance visualization:

sns.countplot(x="default_flag", data=dfcopy)
plt.title("Default distribution (0 = paid, 1 = default)")
plt.tight_layout()
plt.show()
```

```python
#Distribution comparison: loan amount by default
sns.boxplot(x= "default_flag", y = "loan_amnt", data = dfcopy)
plt.title("Loan amount by default")
plt.tight_layout()
plt.show
```

```python
to_summarize = [c for c in ["loan_amnt","funded_amnt","int_rate_clean","annual_inc","dti","fico_mid","revol_util_clean"] if c in dfcopy.columns]

print(dfcopy.groupby("default_flag")[to_summarize].mean(numeric_only=True).T)
```

```python
#Default rate by grade
if "grade" in dfcopy.columns:
    grade_default = dfcopy.pivot_table(index="grade", values = "default_flag", aggfunc="mean")
    print((grade_default * 100).round(2).sort_index())

    sns.barplot(x=grade_default.index, y = (grade_default["default_flag"]* 100).values)
    plt.ylabel("Default rate (%)")
    plt.title("Default rate by grade")
    plt.tight_layout()
    plt.show()
```

```python
#Default rate by term 
if "term_months" in dfcopy.columns:
    term_default = dfcopy.pivot_table(index="term_months", values="default_flag", aggfunc="mean")
    print((term_default * 100).round(2).sort_index())

    sns.barplot(x= term_default.index.astype(int), y =(term_default["default_flag"]*100).values)
    plt.xlabel("Term(months)")
    plt.ylabel("Default rate (%)")
    plt.title("Default rate by term")
    plt.tight_layout()
    plt.show()
```

```python
# Building a correlation matrix of numeric columns plus the target.
corr_cols = [c for c in num_cols if c in dfcopy.columns] + ["default_flag"]
corr = dfcopy[corr_cols].corr(numeric_only=True)

print(corr["default_flag"].sort_values(ascending=False))

plt.figure(figsize=(9,7))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation heatmap")
plt.tight_layout()
plt.show()
```
